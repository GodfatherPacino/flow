## read edges from the file&the k-shortest

```python
import networkx as nx
from itertools import islice

#G=nx.Graph()
G = nx.read_edgelist('edges_505', create_using=nx.DiGraph(),nodetype=int, data=(('mode',int), ('cap',float),('flow',float),('len',float),('disu',float)))
#print(G.edges.data())
print(list(G.edges.data())[0])

```

```python
def k_shortest_paths(G, source, target, k, weight=None):
    return list(islice(nx.shortest_simple_paths(G, source, target, weight='disu'), k))
for path in k_shortest_paths(G, 0, 58, 3):
    print (len(path))
    print (path)
    sum=float(0)
    for i in range(len(path)-1):
        weight=G.edges[path[i],path[i+1]]['disu']
        sum=sum+weight
    print (sum)
```

```python
import re
fh=open("expandnode2_forweight.txt", "r")
srs=fh.readlines()
for sr in srs:
    sr_1=re.split('\s', sr)
    #print(sr_1)
    a=int(sr_1[0])
    b=float(sr_1[1])
    c=float(sr_1[2])
    G.add_node(a, Loc=b, Tra=c)
    

print(G.nodes[0]["Loc"])
print(G.nodes.data())
```
